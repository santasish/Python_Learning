#### Matplotlib

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

### Lineplot

```python
plt.plot(x,y)
plt.show()
```

Adding the titles and labels..

```python
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

Changing line aesthetics...

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.show()
```

Adding 2 lines in the same plot

```python
x=np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=3)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

#### Adding sub-plots

```python
plt.subplot(1,2,1)    # 1 row, 2 columns, 1 is index of the subplot
plt.plot(x,y1,color='g',linestyle=':',linewidth=4)

plt.subplot(1,2,2)    # 1 row, 2 columns, 2 is the index of the subplot
plt.plot(x,y2,color='r',linestyle='-',linewidth=2)

plt.show()
```

### Barplot

```python
student = {'Harry':86,'Naseem':90,'Murad':75,'Bacon':78}
```

```python
names = list(student.keys())
names
```

```python
marks = list(student.values())
marks
```

```python
plt.bar(names,marks)
plt.title('Distribution of Student Marks')
plt.xlabel("names")
plt.ylabel("marks")        
plt.show()
```

### Horizontal Bar Plot

```python
plt.barh(names,marks,color='g')
plt.title("Bar Plot")
plt.xlabel('names')
plt.ylabel('marks')
plt.grid(True)
plt.show()
```

### Scatter Plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]

plt.scatter(x,a)
plt.show()
```

Changing mark aesthetics..

```python
plt.scatter(x,a,marker='*',c='r',s=100)
plt.show()
```

Adding 2 markers in the same plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[6,2,1,5,3,8,4,7,0]

plt.scatter(x,a,marker='*',c='g',s=80)
plt.scatter(x,b,marker='.',c='r',s=100)
plt.show()
```

Adding sub-plots

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c='g',s=80)

plt.subplot(1,2,2)
plt.scatter(x,b,marker='.',c='r',s=80)

plt.show()
```

### Pie Chart

```python
BI_Tools = ['tableau','powerbi','alteryx','Qclicksense']
popularity=[80,78,28,16]
```

```python
plt.pie(popularity,labels=BI_Tools)
plt.show()
```

Changing aesthetics...

```python
plt.pie(popularity,labels=BI_Tools,autopct='%0.1f%%',colors=['violet','grey','blue','black'])
plt.show()
```

### Donut Chart

```python
plt.pie(popularity,labels=BI_Tools,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```

## Seaborn

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.shape
```

### Line plot

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

Grouping data with hue...

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue="event")
plt.show()
```

Adding styles...

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue="event",style='event')
plt.show()
```

Adding markers...

```python
sns.lineplot(x='timepoint',y='signal',markers=True,hue="event",style='event',data=fmri)
plt.show()
```
